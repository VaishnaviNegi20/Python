```python
#overloading of greater than (gt) comparison operator
class Person:
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def __gt__(self,other): #this is define for the Person class
        if self.age > other.age: # it's working fine, and it'll call the __gt__ define for the int class to compare because the p1 and p2 age are of int datatype
            return True
        else:
            return False
        
        
p1 = Person("akash",29)
p2 = Person("gautam",55 )
if p1>p2:  # here the def __gt__(self,other) method which we define, would be called
    print(f"{p1.name} is going to pay the bill")
else:
    print(f"{p2.name} is going to pay the bill")
    
```
