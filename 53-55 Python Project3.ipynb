# Hangman Game

```python
import random

import hangman_stages
import word_file

word_selected = random.choice(word_file.words)
list1 = list(word_selected)
length = len(word_selected)

list2 = list1.copy()

for i in range(length):
    list2[i] = "_"

total_guess = 6
while True:
    user_guess = input("guess a letter : ").lower()
   
    if user_guess in list1:
        for i in range(length):
            if list1[i] == user_guess:
                list2[i] = user_guess
        print(list2)
        if list2 == list1:
            print("You win")
            break
    else:
        total_guess-=1
        print(f"{total_guess} lives remain\n")
        if total_guess==0:
            print("You lose")
            break
    print(hangman_stages.stages[total_guess])
```
